Project Objective:

Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
sc = df['Species'].value_counts()
sc.plot(kind = 'bar')
plt.title("Total Number of Iries Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df['SepalLengthCm'])
plt.title("Breakdown of Sepal Length")
plt.xlabel("Sepal length")
plt.ylabel("Frequency")
```

```python
plt.scatter(df['SepalLengthCm'], df['PetalLengthCm'])
plt.title("Sepal Length v/s Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
```

Root Discovery:

- The dataset shows a highly balanced class distribution across Iris species.

- There is a relationship between sepal length and petal length.

- By visualizing the dataset, it makes easier to identify trends, distribution and pattern that might not seem obvious through raw data.
